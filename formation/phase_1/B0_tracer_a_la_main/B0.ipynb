## Section 1 — Rappels synthétiques

### Qu'est-ce que le traçage ?

Tracer = jouer le rôle de l'ordinateur :
1. Lire chaque ligne dans l'ordre
2. Remplacer les variables par leur valeur courante
3. Mettre à jour le tableau
4. Suivre les conditions (True/False) et les boucles

### Rappel du module précédent (A3)

Tu sais que `boucle_de_jeu()` appelle `executer_mouvement()` à chaque
déplacement valide. Dans ce module, on plonge à l'intérieur de cette fonction.

### Les 4 structures modifiées par executer_mouvement()

| Structure | Ce qui change |
|-----------|---------------|
| `drone['col']`, `drone['lig']` | nouvelles coordonnées |
| `etat['grille'][lig][col]` | symbole effacé + placé |
| `drone['batterie']` | décrémenté du coût |
| valeur de retour | log de l'action |

### Étape 2 — Vérifier par l'exécution

Exécute la vraie fonction et compare avec ton traçage.

````python
import copy

def executer_mouvement(etat, drone_id, col_dest, lig_dest):
    """Déplace le drone et met à jour toutes les structures."""
    drone = etat['drones'][drone_id]
    col_avant = drone['col']
    lig_avant = drone['lig']
    drone['col'] = col_dest
    drone['lig'] = lig_dest
    etat['grille'][lig_avant][col_avant] = ' . '
    etat['grille'][lig_dest][col_dest]   = ' D '
    cout = 1
    if (col_dest, lig_dest) in etat['zones_x']:
        cout += 2
    drone['batterie'] -= cout
    return f"{drone_id}:({col_avant},{lig_avant})->({col_dest},{lig_dest}), batterie={drone['batterie']}"

# Réinitialiser l'état
grille2 = [
    [' . ', ' . ', ' . ', ' . '],
    [' . ', ' . ', ' . ', ' . '],
    [' . ', ' D ', ' . ', ' . '],
    [' . ', ' . ', ' . ', ' . '],
]
drone2 = {'id': 'D1', 'col': 1, 'lig': 2, 'batterie': 10, 'batterie_max': 20}
etat2 = {'drones': {'D1': drone2}, 'grille': grille2, 'zones_x': set()}

log = executer_mouvement(etat2, 'D1', 2, 2)

print('=== RÉSULTAT APRÈS MOUVEMENT ===')
print('Log :', log)
print(f"Drone D1 : col={drone2['col']}, lig={drone2['lig']}, batterie={drone2['batterie']}")
print('Grille :')
for i, ligne in enumerate(grille2):
    print(f'  lig {i} : {ligne}')

# Vérifications
assert drone2['col'] == 2, 'col doit valoir 2'
assert drone2['lig'] == 2, 'lig doit valoir 2'
assert drone2['batterie'] == 9, 'batterie doit valoir 9'
assert grille2[2][1] == ' . ', 'ancienne case doit être vide'
assert grille2[2][2] == ' D ', 'nouvelle case doit avoir le drone'
print('\n✅ Toutes les vérifications passent !')
````

### Étape 4 — Traçage avec debug print

Ajoute des `print()` stratégiques dans `executer_mouvement()`
pour afficher chaque variable clé après chaque modification.

````python
def executer_mouvement_debug(etat, drone_id, col_dest, lig_dest):
    """Version avec print de debug pour visualiser le traçage."""
    drone = etat['drones'][drone_id]
    print(f'[DEBUG] avant : col={drone["col"]}, lig={drone["lig"]}, batterie={drone["batterie"]}')

    col_avant = drone['col']
    lig_avant = drone['lig']
    drone['col'] = col_dest
    drone['lig'] = lig_dest
    print(f'[DEBUG] coords mises à jour : ({col_avant},{lig_avant}) → ({col_dest},{lig_dest})')

    etat['grille'][lig_avant][col_avant] = ' . '
    etat['grille'][lig_dest][col_dest]   = ' D '
    print(f'[DEBUG] grille[{lig_avant}][{col_avant}] effacée, grille[{lig_dest}][{col_dest}] = D')

    cout = 1
    if (col_dest, lig_dest) in etat['zones_x']:
        cout += 2
        print(f'[DEBUG] zone X détectée sur ({col_dest},{lig_dest}) → coût = {cout}')
    else:
        print(f'[DEBUG] pas de zone X → coût = {cout}')

    drone['batterie'] -= cout
    print(f'[DEBUG] après : batterie = {drone["batterie"]}')

    return f"{drone_id}:({col_avant},{lig_avant})->({col_dest},{lig_dest}), batterie={drone['batterie']}"

# Test
grille4 = [[' . ']*4 for _ in range(4)]
grille4[2][1] = ' D '
drone4 = {'id': 'D1', 'col': 1, 'lig': 2, 'batterie': 10, 'batterie_max': 20}
etat4 = {'drones': {'D1': drone4}, 'grille': grille4, 'zones_x': {(2, 2)}}

print('=== TRACE AVEC DEBUG ===')
log4 = executer_mouvement_debug(etat4, 'D1', 2, 2)
print('Log final :', log4)
````
