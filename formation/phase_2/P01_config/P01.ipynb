### Principe DRY

- **Sans config externe** : `12` apparaît à 10 endroits dans le code
- **Avec config externe** : `GRILLE_TAILLE = config['GRILLE_TAILLE']` — un seul endroit

### Rappel A2

`json.load()` retourne un `dict` Python. Tu accèdes aux valeurs par clé : `donnees['GRILLE_TAILLE']`.
