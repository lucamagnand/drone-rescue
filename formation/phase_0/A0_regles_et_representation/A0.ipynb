### Convention de coordonnées

- Colonnes : lettres `A` à `L` (index 0 à 11)
- Lignes : chiffres `1` à `12` (index 0 à 11)
- Accès Python : `grille[lig][col]` (ligne d'abord, colonne ensuite)
- Coordonnée humaine `C4` = colonne 2 (C), ligne 3 (4-1=3)
