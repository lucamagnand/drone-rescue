### Qu'est-ce que le pseudo-code ?

Le pseudo-code est un texte structuré entre le français et le code :
- **Séquentiel** : les étapes sont dans l'ordre d'exécution
- **Atomique** : chaque étape fait une seule chose
- **Non ambigu** : on sait ce qui entre et ce qui sort

| Trop vague | Bon niveau | Trop détaillé |
|------------|------------|---------------|
| Gérer le tour | Valider le mouvement du drone | Vérifier si `col >= 0 and col < 12` |
