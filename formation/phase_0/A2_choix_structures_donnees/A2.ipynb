## Section 1 — Rappels synthétiques

### Les 4 structures fondamentales

| Structure | Syntaxe       | Accès          | Modifiable | Doublons |
|-----------|---------------|----------------|------------|----------|
| `list`    | `[a, b, c]`   | indice entier  | oui        | oui      |
| `dict`    | `{cle: val}`  | clé nommée     | oui        | clés uniques |
| `set`     | `{a, b, c}`   | appartenance   | oui        | non      |
| `tuple`   | `(a, b, c)`   | indice entier  | **non**    | oui      |

### Règle de décision rapide
- Accès par **nom** → `dict`
- **Appartenance** rapide (O(1)) → `set`
- **Séquence ordonnée** avec doublons → `list`
- **Coordonnée immuable** (hashable) → `tuple`

### Étape 2 — Tester la différence list vs set

Voyons concrètement pourquoi un `set` est préférable à une `list`
pour vérifier si une position est occupée.

````python
import time

# Simulons 1000 positions occupées
positions_list = [(i % 12, i // 12) for i in range(1000)]
positions_set  = set(positions_list)

cible = (11, 11)  # dernière case

# Mesure avec list
debut = time.perf_counter()
for _ in range(100_000):
    _ = cible in positions_list
duree_list = time.perf_counter() - debut

# Mesure avec set
debut = time.perf_counter()
for _ in range(100_000):
    _ = cible in positions_set
duree_set = time.perf_counter() - debut

print(f'Avec list : {duree_list:.4f}s')
print(f'Avec set  : {duree_set:.4f}s')
print(f'Ratio     : set est ~{duree_list/duree_set:.0f}x plus rapide')
````
