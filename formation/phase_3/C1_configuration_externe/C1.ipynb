## Section 1 — Rappels synthétiques

### Principe
Une **valeur magique** = un littéral (`12`, `10`, `0.5`) écrit directement dans le code sans nom explicite.
Une **constante nommée** = une variable en MAJUSCULES (`GRILLE_TAILLE`, `BATTERIE_MAX`) chargée depuis un fichier externe.

### Pourquoi externaliser ?
- Modifier la taille de la grille = changer **1 valeur** dans `config.json`
- Sans externalisation = chercher `12` dans 6 fichiers et 600 lignes

### Structure de config.json (Drone Rescue)
```json
{
  "grille_taille": 12,
  "batterie_max": 10,
  "tours_max": 30,
  "cout_deplacement": 1,
  "cout_zone_x": 2,
  "recharge_stationnaire": 3
}
```

### Rappel du module précédent (C0)
`config.py` est le seul module qui lit `config.json`. Tous les autres importent depuis `config.py`.
